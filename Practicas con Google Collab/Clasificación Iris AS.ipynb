```python
# Se usa para cargar el dataset de las flores iris, medidas de los petalos y los capullos
from sklearn.datasets import load_iris

DataIris = load_iris()
DataIris

#
from sklearn.model_selection import train_test_split

#
from sklearn.ensemble import RandomForestClassifier

#
from sklearn.metrics import accuracy_score

#
## test_size = 0.3
## random_state = 47

# Obtenemos los datos y etiquetas
x = DataIris.data
y = DataIris.target

Data_Train, Data_Test, Target_Train, Target_Test = train_test_split(x,y,test_size = 0.5, random_state = 47)
Target_Test

print(" ----------------------------------------------------------------------------")

print("\n Tamaño de Datos de Dataset para Entrenamiento: ", Data_Train.shape)
print(" Tamaño de Datos de Dataset para Pruebas", Data_Test.shape)

print("\n Tamaño de Etiquetas para Entrenamiento", Target_Train.shape)
print(" Tamaño de Etiqueras para Pruebas" ,Target_Test.shape)

print("\n----------------------------------------------------------------------------")

# Crear el modelo que va a clasificar
# Random forest, este algoritmo esta basado en bosques aleatorios (n de arboles)
NumeroDeArboles = 420

Modelo = RandomForestClassifier(NumeroDeArboles, random_state=47)
print("\n", Modelo)

print("\n----------------------------------------------------------------------------")

#
Modelo.fit(Data_Train,Target_Train)

#
Data_Pred = Modelo.predict(Data_Test)
print("\n", Data_Pred)

print("\n----------------------------------------------------------------------------")

#
Precision = accuracy_score(Target_Test,Data_Pred)
print(f"\n Precision del Modelo: {Precision * 100: 0.2f}")

print("\n----------------------------------------------------------------------------")
```
