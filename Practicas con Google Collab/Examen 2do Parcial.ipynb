# Problema 1

**Generar un nuevo dataset de vehículos y sus características.** Los datos incluirán información sobre vehículos como "Marca", "Modelo", "Año", "Precio", "Kilometraje", "Fecha de Registro". 100 registros

* Eliminar duplicados.
* Tratar valores nulos: rellenar o eliminar, dependiendo del contexto.
* Corregir tipos de datos.
* Eliminar espacios en blanco adicionales.
* Ordenar el DataFrame por el precio.
* Guardar el DataFrame limpio en un archivo CSV.

```python
# Imports primero
import pandas as pd
import numpy as np
import random
import google.colab.drive as drive

drive.mount("/content/drive",force_remount=True)
TotalAutos = 100

Data = {
    "Nombre del Modelo":[f"Auto {i+1}" for i in range(TotalAutos)],
    "Marca":[f"Marca {random.randint(1,10)}" for i in range(TotalAutos)],
    "Año":[random.randint(1950,2020) for i in range(TotalAutos)],
    "Precio":[random.randint(20000,150000) for i in range(TotalAutos)],
    "Kilometraje":[random.randint(20000,150000) for i in range(TotalAutos)],
    "Fecha de Registro":[pd.Timestamp("1-1-2024") + pd.DateOffset(days=random.randint(0,365)) for i in range(TotalAutos)]
}

df = pd.DataFrame(Data)
print(df.head())

print("-----------------------------------------------------------------------------")

# Eliminar Duplicados
df.drop_duplicates(inplace=True)
print("Duplicados Eliminados")

print("-----------------------------------------------------------------------------")

# Ordenar el DataFrame por Precio
df_sorted = df.sort_values(by='Precio')

# Mostrar el DataFrame Ordenado
print(df_sorted)

print("-----------------------------------------------------------------------------")

# Generamos el archivo CSV
df.to_csv('/content/drive/MyDrive/Colab Notebooks/Examen 2 Ej 1.csv', index= False)
print("Archivo Generado")
```

# Problema 3

**Generar un nuevo dataset de clientes en una tienda.** Los datos incluirán información sobre clientes como "ID Cliente", "Nombre", "Edad", "Correo Electrónico", "Fecha de Registro", "Total
Gastado".

* Eliminar duplicados.
* Tratar valores nulos: rellenar o eliminar, dependiendo del contexto.
* Corregir tipos de datos.
* Eliminar espacios en blanco adicionales.
* Ordenar el DataFrame por el total gastado.
* Guardar el DataFrame limpio en un archivo CSV.

```python
# Imports primero
import pandas as pd
import numpy as np
import random
import google.colab.drive as drive

drive.mount("/content/drive",force_remount=True)
TotalClientes = 100

Data = {
 "ID del Cliente":range(1001,1101),
 "Nombre del Cliente":[f"Cliente {i+1}" for i in range(TotalClientes)],
 "Edad":[random.randint(19,50) for i in range(TotalClientes)],
 "Correo Electronico":[f"Correo{i+1}@gmail.com" for i in range(TotalClientes)],
 "Fecha de Registro":[pd.Timestamp("1-1-2024") + pd.DateOffset(days=random.randint(0,365)) for i in range(TotalClientes)],
 "Total Gastado":[random.randint(20,5000) for i in range(TotalClientes)]
}

df = pd.DataFrame(Data)
print(df.head())

print("-----------------------------------------------------------------------------")

# Eliminar Duplicados
df.drop_duplicates(inplace=True)
print("Duplicados Eliminados")

print("-----------------------------------------------------------------------------")

# Ordenar por total gastado
df_sorted = df.sort_values(by='Total Gastado')

# Mostrar el DataFrame ordenado por total gastado
print(df_sorted)

print("-----------------------------------------------------------------------------")

# Generamos el archivo CSV
df.to_csv('/content/drive/MyDrive/Colab Notebooks/Examen 2 Ej 3.csv', index= False)
print("Archivo Generado")
```
